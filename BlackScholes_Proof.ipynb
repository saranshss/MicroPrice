##Black Scholes Proof##

$$dS = \mu S dt + \sigma S dW_t^P$$
$$dB = rB dt$$

$$C(t,S_t); C(T,S_T) = \phi(S_T) = (S_T - K)^+$$

From applying Ito's Lemma on C, we get 

$$dC =C_tdt + C_sdS + \frac{1}{2} C_{ss} (dS)^2$$

Furthurmore, we construct a portfolio V from 1 Call option and $\alpha$ stocks

$$V = C + \alpha S$$

Hence, we have from Ito's lemma,

$$dV = dC + \alpha dS$$

Plugging the value of dC and dS in dV, we get,

$$dV = C_tdt + C_sdS + \frac{1}{2} C_{ss} (dS)^2 + \alpha dS$$

$$dV = C_tdt + (C_s + \alpha)dS + \frac{1}{2} C_{ss} (dS)^2$$

$$dV = C_tdt + (C_s + \alpha)(\mu S dt + \sigma S dW_t^P) + \frac{1}{2} C_{ss} (\mu S dt + \sigma S dW_t^P)^2$$

$$dV = (C_t + \mu S(C_s + \alpha) + \frac{1}{2} C_{ss}\sigma^2 S^2)dt + \sigma S (C_s + \frac{1}{2}\alpha) dW^p$$

$$dV = (C_t + \mu S C_s + \mu S\alpha + \frac{1}{2} C_{ss}\sigma^2 S^2)dt + (\sigma S C_s + \sigma S \alpha) dW^p$$

To make the portfolio risk free, the risk part, i.e., the coefficient of the $dW^p$ term in $dV$ should go to 0, 
$$\alpha = -C_s$$

$$\therefore V = C - C_s S$$ 

$$dV = (C_t + \frac{1}{2} \sigma^2 S^2 C_{ss})dt\text{;   No risk in portfolio}$$

From no arbitrage condition, we can fix the return of the portfolio and the bond as equal, hence

$$dV = rV dt$$

Plugging the value of V from above, we get,

$$dV = r(C - S C_s) dt = (C_t + \frac{1}{2} \sigma^2 S^2 C_{ss})dt$$

From this, we get the Black Scholes equation,

$$C_t + rSC_s + \frac{1}{2} \sigma^2 S^2 C_{ss} - rC = 0$$

Looking at this equation, we can write $dS$ as,

$$dS = rSdt + \sigma S dW^Q$$

Where $dW^Q = dW^P + \frac{\mu - r}{\sigma}dt$, and $\mu$ may or may not be equal to $r$

Looking at Black Scholes eqn, we can say that the price of the call option is a matingale.

Now from Feynman Kac equation, we have,

$$e^{-rt}C(t,S_t) = E_t[e^{-rT}C(T,S_T)]$$

$$C(t,S_t) = \frac{E_t[e^{-rT}C(T,S_T)]}{e^{-rt}}$$

$$C(t,S_t) = e^{-r(T-t)} E_t[C(T,S_T)]$$

Where $$C(T,S_T) = (S_T - K)^+$$

Substituting $\tau = T - t$ and $S = e^y$,
$$log S = y$$
$$d(log S) = (r - \frac{\sigma^2}{2})dt + \sigma dW_t$$

Integrating this from t to T,

$$\int_t^T d(log S) = \int_t^T (r - \frac{\sigma^2}{2})dt + \int_t^T\sigma dW_t$$

$$log S_T - log S_t = (r - \frac{\sigma^2}{2}) \tau + \sigma(W_T - W_t)$$

$$y_T - y_t = (r - \frac{\sigma^2}{2}) \tau + \sigma(W_T - W_t)$$

$$ y_T \sim N(y_t + (r - \frac{\sigma^2}{2}) \tau,\sigma^2)$$

From the definition of nomal distribution, we know that the pdf of a normally distributed random variable x is

$$f(x) = \frac{1}{\sqrt{2\pi\sigma_x^2}} exp\bigg(-\frac{(x_T - x_t)^2}{2\sigma_x^2}\bigg)$$

And $E[x] = \int_{-\infty}^{\infty}mf_x(m) dm$

So we have

$$C(t,S_t) = e^{-r\tau}E_t[(S_T - K)^+] = e^{-r\tau}E_t[(e^{y_T} - K)^+]$$

$$C(t,S_t) = e^{-r\tau} \int_{-\infty}^{\infty} (e^{y_T} - K)^+f(y_T) dy_T$$

Taking $y^* = log K$, $e^{y^*} = K$,

$$C(t,S_t) = e^{-r\tau} \int_{y^*}^{\infty} (e^{y_T} - K) f(y_T) dy_T$$

$$C(t,S_t) = e^{-r\tau} \int_{log K}^{\infty} (e^{y_T} - K) f(y_T) dy_T$$

$$C(t,S_t) = e^{-r\tau} \int_{log K}^{\infty} e^{y_T} f(y_T) dy_T - e^{-r\tau} \int_{log K}^{\infty} K f(y_T) dy_T$$

Taking the first part of the above as A and the second part as B, and tackling solving B first

$$ B =  e^{-r\tau} \int_{log K}^{\infty} K f(y_T) dy_T$$

$$ B =  K e^{-r\tau} \int_{log K}^{\infty} f(y_T) dy_T$$

$$ B =  K e^{-r\tau} \int_{-\infty}^{-log K} f(y_T) dy_T$$

We know that $y_T > log K$, hence,

$$ y_t + \bigg(r - \frac{\sigma^2}{2}\tau + \sigma (W_T - W_t) \bigg) > log K $$

$$ log S_t + \bigg(r - \frac{\sigma^2}{2}\tau + \sigma (W_T - W_t) \bigg) > log K $$

$$  W_T - W_t  > \frac{log (K/S_t) - \bigg(r - \frac{\sigma^2}{2}\tau\bigg)}{\sigma}$$

To make LHS a standard normal distribution, we divide both sides by $\sqrt{\tau}$,

$$  \frac{W_T - W_t}{\sqrt {\tau}}  > \frac{log (K/S_t) - \bigg(r - \frac{\sigma^2}{2}\tau\bigg)}{\sigma\sqrt{\tau}}$$

From the property of CDF, that $CDF(X) = CDF(-X)$, we get,

$$  \frac{W_T - W_t}{\sqrt {\tau}}  < \frac{log (S_t/K) + \bigg(r - \frac{\sigma^2}{2}\tau\bigg)}{\sigma\sqrt{\tau}}$$

The LHS here is the CDF if the integral part of B, hence we get,

$$B = K e^{-r\tau} Pr \bigg(\phi < \frac{log (S_t/K) + (r - \frac{\sigma^2}{2})\tau}{\sigma \sqrt{\tau}}\bigg)$$

$$ B = K e^{-r\tau} N(d_2)$$

$N(d_2)$ is the probability of the call option expiring in the money.

$$A = e^{-r\tau} \int_{log K}^{\infty} e^{y_T} f(y_T) dy_T$$ 

$$A = \frac{e^{-r\tau}}{\sqrt{2\pi \sigma_y^2}} \int_{log K}^{\infty} exp\bigg(y_T - \frac{(y_T - \mu_y)^2}{2\sigma_y^2} \bigg) dy_T$$ 

Solving the exponent, we get 

$$ = y_T - \frac{(y_T - \mu_y)^2}{2\sigma_y^2} = \frac{2\sigma_y^2y_T - (y_T - \mu_y)^2}{2\sigma_y^2}$$
